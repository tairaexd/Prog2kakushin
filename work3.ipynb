```python
import numpy as np
import cv2, sys
from google.colab.patches import cv2_imshow
import random
import time

pt=['canal.jpg','desert.jpg','dog.jpg','eagle.jpg','flower.jpg','fox.jpg','ichidai.jpg','library.jpg','parrot.jpg','turtle.jpg']
imgw = np.ones((1000,300,3), np.uint8)
imgw=imgw*255
for i in range(1):
  print('5秒で覚えてください.')
  imge=random.choice(pt)
  img=cv2.imread(imge)

  if img is None:
    sys.exit('エラー: ファイルを開けませんでした')

  cv2_imshow(img)

  for x in [5,4,3,2,1]:
    time.sleep(1)
    print(str(x)+'秒前....')
  cv2_imshow(imgw)

  img2 = cv2.blur(img,(3,3))
  img3 = cv2.detailEnhance(img, sigma_s=10, sigma_r=0.15)
  img4 = cv2.edgePreservingFilter(img, flags=1, sigma_s=120, sigma_r=0.5)
  data=[img,img2,img3,img4]
  data2=[0,1,2,3]
  random.shuffle(data2)


  for j in range(4):
    print('No.'+str(j+1))
    if data2[j]==0:
      ans=j+1
    cv2_imshow(data[data2[j]])

  q=int(input('元の画像はどれですか？(数字)'))
  if q==ans:
    print('正解！')








```
