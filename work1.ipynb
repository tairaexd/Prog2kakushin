```python
import pandas as pd
import csv
print('「表」について学ぼう!')
tdf=pd.DataFrame([['メロン','茨城県','みどり'],['ブドウ','山梨県','むらさき'],['リンゴ','青森県','あか']])
tdf.index=['第一位','第二位','第三位']
tdf.columns=['果物名','名産地','色']
print('下にあるのは2023年の果物の人気ランキングTOP3の「表」です.')
print('「例えば,三番目に人気のリンゴの名産地は青森県だ」のように,パッと見てすぐにわかるのが「表」なのです.これがすべて文章で書かれていたら,わかりづらいですよね.')
print('横の並びを「行」,縦の並びを「列」といい,「果実名」,「名産地」,「色」を 列名,「第一位」,「第二位」,「第三位」を行名と呼びます.')
tdf
```

```python
print('表についての分かっているかの簡単なクイズに答えてみてください.')
print('全問正解でクリアです!')
count=0
while count<3:
  count=0
  ans1=input('[1]人気第三位の果物は?:')
  if ans1==tdf.at['第三位','果物名']:
    print('正解です!')
    count+=1
  else:
    print('間違いです,正解は「リンゴ」です.')
  ans2=input('[2]ブドウの名産地は?:')
  if ans2==tdf.at['第二位','名産地']:
    print('正解です!')
    count+=1
  else:
    print('間違いです,正解は「山梨県」です')
  ans3=input('二行目の三列目は?:')
  if ans3==tdf.at['第二位','色']:
    print('正解です!')
    count+=1
  else:
    print('間違いです,正解は「むらさき」です.')
  if count<3:
    print('もう一度挑戦しましょう!')
print('全問正解です!')


```

```python
import pandas as pd
import csv
print('表について学べたところで,自分で簡単な表を作成してみましょう!')
rs=int(input('何行の表を作成しますか(数字で列名は含まない)：'))
cs=int(input('何列の表を作成しますか(数字で行名は含まない)：'))
data=[]
minir=[]
nr=[]
nc=[]
for rn in range(rs):
  minir=[]
  for cn in range(cs):
    v=input(str(rn+1)+'行目の'+str(cn+1)+'列目は?')
    minir.append(v)
    if cn==cs-1:
      data.append(minir)
df=pd.DataFrame(data)
for rn in range(rs):
  a=input(str(rn+1)+'行目の行名は?')
  nr.append(a)
for cn in range(cs):
  b=input(str(cn+1)+'列目の列名は?')
  nc.append(b)
df.index=nr
df.columns=nc
print('あなたが作成した表はこちらです.イメージしたと通りに作れましたか?')
print('できるまで何度もやってみよう!')
df
```
