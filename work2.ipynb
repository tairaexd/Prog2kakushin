```python
#～乗の関数
import numpy as np
import matplotlib.pyplot as plt
n=int(input('何次関数?'))
k=[]
for i in range(n+1):
  if i==0:
    kei=int(input('定数項は?'))
    k.append(kei)
  else:
    kei=int(input(str(i)+'乗の項の係数は？'))
    k.append(kei)
xs=int(input('定義域の始点は？(一番左)'))
xf=int(input('定義域の終点は？(一番右)'))
x=np.linspace(xs,xf,(xf-xs)*10)
y=0
for l in range(n+1):
  y+=k[l]*x**l
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
print('グラフは上のような形になります')
```

```python
#指数関数
import numpy as np
import matplotlib.pyplot as plt
n=int(input('項数は？'))
tt=[]
ts=[]
tk=[]
for i in range(n):
  e1=int(input(str(i+1)+'つめの項の底は？'))
  e2=int(input('またその指数であるxの係数は？'))
  e3=int(input('またその項の符号は？(プラスなら「+1」マイナスなら「-1」)'))
  tt.append(e1)
  ts.append(e2)
  tk.append(e3)
xs=int(input('定義域の始点は？(一番左)'))
xf=int(input('定義域の終点は？(一番右)'))
x=np.linspace(xs,xf,(xf-xs)*10)
y=0
for k in range(n):
  y+=tk[k]*(tt[k]**(ts[k]*x))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
print('グラフは上のような形になります')
```
